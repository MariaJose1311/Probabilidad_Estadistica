# EJERCICIO 1 – Estadística Descriptiva (Media, Mediana, Varianza, Desviación estándar)
# -------------------------------------------------------------------
Instrucciones: Crea un conjunto de datos simulado de ingresos de 100 personas
y calcula desde cero:
 - Media
 - Mediana
 - Varianza
 - Desviación estándar
 - Coeficiente de variación

```python
# Funciones
# Definimos una función que calcula la media de un array de números.
def media(numarray):
    return sum(numarray) / len(numarray)

# Definimos una función que calcula la mediana de un array de números.
def mediana(numarray):
    # Primero, ordenamos el array
    numarray = sorted(numarray)
    # Luego, calculamos la mediana
    n = len(numarray)
    # Si el número de elementos es par, la mediana es el promedio de los dos del medio
    if n % 2 == 0:
        return (numarray[n // 2 - 1] + numarray[n // 2]) / 2
    else:
    # Si el número de elementos es impar, la mediana es el del medio
        return numarray[n // 2]
    
# Definimos una función que calcula la varianza de un array de números.
def varianza(numarray):
    # Primero, calculamos la media
    media_numarray = media(numarray)
    # Luego, calculamos la varianza
    return sum((x - media_numarray) ** 2 for x in numarray) / len(numarray)

# Definimos una función que calcula la desviación estándar de un array de números.
def desviacion_estandar(numarray):
    # Primero, calculamos la varianza
    var_numarray = varianza(numarray)
    # Luego, calculamos la desviación estándar
    return var_numarray ** 0.5

# Definimos una función que calcula el coeficiente de variación de un array de números.
def coeficiente_variacion(numarray):
    # Primero, calculamos la desviación estándar
    des_est = desviacion_estandar(numarray)
    # Luego, calculamos la media
    media_numarray = media(numarray)
    # Finalmente, calculamos el coeficiente de variación
    return des_est / media_numarray
```

```python
import numpy as np
```

```python
# Queremos crear un conjunto de datos simulado de ingresos de 100 personas y calcula desde 0
np.random.seed(42)  # Para reproducibilidad
ingresos = np.random.normal(loc=50000, scale=1000, size=100)
```

```python
# Cálculo

# Calculamos la media, mediana, varianza, desviación estándar y coeficiente de variación
media_ingresos = media(ingresos)
mediana_ingresos = mediana(ingresos)
varianza_ingresos = varianza(ingresos)
desviacion_ingresos = desviacion_estandar(ingresos)
coeficiente_variacion_ingresos = coeficiente_variacion(ingresos)

# Imprimimos los resultados
print(f"Media de ingresos: {media_ingresos:.2f}")
print(f"Mediana de ingresos: {mediana_ingresos:.2f}")
print(f"Varianza de ingresos: {varianza_ingresos:.2f}")
print(f"Desviación estándar de ingresos: {desviacion_ingresos:.2f}")
print(f"Coeficiente de variación de ingresos: {coeficiente_variacion_ingresos:.2f}")
```

# EJERCICIO 2 – Z-score y Detección de Outliers
# -------------------------------------------------------------------
Instrucciones: Utiliza el conjunto de ingresos del ejercicio anterior
y calcula el z-score para cada valor. Luego identifica los valores atípicos
(aquellos con z-score mayor a 3 o menor a -3)

```python
# Definimos la función que calcula el z-score
def z_score(numarray):
    # Primero, calculamos la media
    media_numarray = media(numarray)
    # Luego, calculamos la desviación estándar
    des_est = desviacion_estandar(numarray)
    # Finalmente, calculamos el z-score
    return [(x - media_numarray) / des_est for x in numarray]

# Calculamos el z-score de los ingresos
z_scores_ingresos = z_score(ingresos)

# Imprimimos los z-scores
print("Z-scores de los ingresos:")
for i, z in enumerate(z_scores_ingresos):
    print(f"Persona {i + 1}: {z:.2f}")
```

```python
# Vamos a identificar los ingresos atípicos (aquellos con z-score mayor a 3 o menor a -3)
outliers = [ingresos[i] for i, z in enumerate(z_scores_ingresos) if abs(z) > 3]

# Imprimimos los ingresos atípicos
print("Ingresos atípicos (z-score > 3 o < -3):")
if outliers:
    for i, outlier in enumerate(outliers):
        print(f"Ingreso atípico {i + 1}: {outlier:.2f}")
# Si no hay ingresos atípicos, lo indicamos
else:
    print("No se encontraron ingresos atípicos.")
```

# EJERCICIO 3 – Probabilidad Frecuentista y Simulación
# -------------------------------------------------------------------
Instrucciones: Simula 10.000 lanzamientos de una moneda justa con NumPy.
Luego calcula:
- La frecuencia relativa de caras y cruces
- La probabilidad estimada de obtener al menos 3 caras en 5 lanzamientos consecutivos (repetido 10.000 veces)

```python
# Vamos a simular 10.000 lanzamientos de una moneda justa con Numpy
np.random.seed(42) 
lanzamientos_moneda = np.random.choice(['Cara', 'Cruz'], size=10000)
```

```python
# Definiremos una función que calcula la frecuencia relativa de caras y cruces
def frecuencia_relativa(lanzamientos):
    caras = np.sum(lanzamientos == 'Cara')
    cruces = np.sum(lanzamientos == 'Cruz')
    total = len(lanzamientos)
    return {
        'Cara': caras / total,
        'Cruz': cruces / total
    }

# Definiremos una función que calcula la probabilidad estimada de obtener 
# al menos tres caras en 5 lanzamientos consecutivos (repetido 10000 veces)
experimentos = 10000
caras_en_cinco_lanzamientos = 3
exitos = 0
for i in range(experimentos):
    cinco_lanzamientos = np.random.choice([0,1], size=5)
    if np.sum(cinco_lanzamientos) >= caras_en_cinco_lanzamientos:
        exitos += 1

probabilidad_estimacion = exitos / experimentos
```

```python
# Calculamos la frecuencia relativa de caras y cruces
frecuencia = frecuencia_relativa(lanzamientos_moneda)
# Imprimimos los resultados
print(f"Frecuencia relativa de caras: {frecuencia['Cara']:.2f}")
print(f"Frecuencia relativa de cruces: {frecuencia['Cruz']:.2f}")

# Imprimimos la probabilidad estimada de obtener al menos tres caras en 5 lanzamientos consecutivos
print(f"Probabilidad estimada de obtener al menos {caras_en_cinco_lanzamientos} caras en 5 lanzamientos: {probabilidad_estimacion:.2f}")
```

```python
# Vamos a simular una distribución Bernoulli con p=0.3 y 1000 observaciones
p = 0.3
n = 1000
# Definimos una función que simula una distribución Bernoulli
def simular_bernoulli(p, n):
    return np.random.binomial(1, p, n)

# Simulamos la distribución Bernoulli
bernoulli_data = simular_bernoulli(p, n)
```

```python
# Vamos a silumar una distribución Binomial con n = 10, p = 0.4 y 1.000 observaciones
n_binomial = 10
p_binomial = 0.4
# Definimos una función que simula una distribución Binomial
def simular_binomial(n, p, size):
    return np.random.binomial(n, p, size)

# Simulamos la distribución Binomial
binomial_data = simular_binomial(n_binomial, p_binomial, 1000)
```

```python
# Vamos a simular una distribución de Poisson con lambda = 5 y 1.000 observaciones
lambda_poisson = 5
# Definimos una función que simula una distribución de Poisson
def simular_poisson(lam, size):
    return np.random.poisson(lam, size)

# Simulamos la distribución de Poisson
poisson_data = simular_poisson(lambda_poisson, 1000)
```

Realiza una prueba de hipótesis para evaluar si la proporción real es menor al 90%.

Usa una prueba unilateral y calcula:
- Hipótesis
- Estadístico z
- Valor-p aproximado (utilizando una función propia de CDF de la normal)

```python
# Datos del ejercicio:
n = 400
vacunados = 344
```

```python
# Definimos las hipótesis nula y alternativa

Hipótesis_nula = "proporción sea mayor o igual a 0.9"
Hipótesis_alternativa = "proporción sea menor a 0.9"
```

```python
# Calculamos el estadístico z

# Primero, calculamos la proporción muestral
proporcion_muestra = vacunados/n

# La proporción poblacional bajo la hipótesis nula
p=0.9

# Calculamos el error estándar
error_estandar = math.sqrt(p*(1-p)/n)

# Calculamos el estadístico z
z = (proporcion_muestra - p) / error_estandar

# Imprimimos el estadístico z
print(f"El estadístico z es {z:.2f}")
```

```python
# Vamos a calcular el Valor-p aproximado (utilizando una función propia de CDF de la normal)
# Definimos una función que calcula la CDF de la normal estándar para el valor z dado
def normal_cdf(z):
    return (1 + math.erf(z / math.sqrt(2))) / 2

# Calculamos el Valor-p
valor_p = normal_cdf(z)

# Imprimimos el Valor-p
print(f"El Valor-p aproximado es {valor_p:.7f}")
```

# EJERCICIO 7 – Correlación de Pearson desde cero
# -------------------------------------------------------------------
Instrucciones: Simula dos variables con relación lineal e introduce algo de ruido.
Calcula la correlación de Pearson utilizando únicamente operaciones aritméticas básicas (no .corr())

```python
# Vamos a simular dos variables con relación lineal y algo de ruido

# Usemos numpy para generar datos aleatorios
np.random.seed(42)  

# Simulamos 100 puntos de datos
x = np.random.uniform(0, 10, 100)

# Simulamos una variable dependiente (y=2*x) y una relación lineal con algo de ruido (+np.random.normal(0, 1, 100))
y = 2 * x + np.random.normal(0, 1, 100)

# Vamos a graficar la relación entre las dos variables
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 2))
plt.scatter(x, y, color='blue', alpha=0.5)
plt.title('Relación entre X e Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
# Vamos a calcular la correlación de Pearson entre las dos variables

# Definimos una función que calcula la correlación de Pearson
def correlacion_pearson(x, y):
    # Primero, debemos calcular las medias de x e y
    media_x = media(x)
    media_y = media(y)
    
    # Segundo, calculamos la covarianza
    covarianza = sum((x[i] - media_x) * (y[i] - media_y) for i in range(len(x))) / len(x)
    
    # Tercero, debemos calcular las desviaciones estándar de x e y
    desviacion_estandar_x = desviacion_estandar(x)
    desviacion_estandar_y = desviacion_estandar(y)
    
    # Y ahora si podemos calcular el coeficiente de correlación de Pearson
    return covarianza / (desviacion_estandar_x * desviacion_estandar_y)
```

```python
# Calculamos la correlación de Pearson para nuestras variables x e y
correlacion = correlacion_pearson(x, y)

# Imprimimos la correlación
print(f"La correlación de Pearson entre X e Y es: {correlacion:.2f}")
```

# EJERCICIO 8 – Correlación de Spearman desde cero
# -------------------------------------------------------------------
Instrucciones: Simula dos variables con relación monotónica (no lineal).
Asigna rangos y calcula Spearman usando la fórmula basada en rangos y diferencia de rangos

```python
# Vamos a simular dos variables con relación monotónica (no lineal)
# importamos curve_fit y least_squares de scipy.optimize
from scipy.optimize import curve_fit,least_squares

# Definimos los datos de la simulación
x_datos  = np.linspace(0, 10, 100)
y_exacto = np.exp(-(x_datos-2.8)**2/0.5**2)+np.exp(-(x_datos-4.8)**2/1.5**2)

# Para comparar después, Parámetros en orden de aparición:
#parametros_reales = [2.8, 0.5, 4.8, 1.5]

# Agregamos ruido a los datos simulados
np.random.seed(42)
y_ruido = 0.05 * np.random.normal(size=x_datos.size)
y_datos = y_exacto + y_ruido

# Graficamos los datos medidos y la curva exacta
plt.figure()
plt.plot(x_datos,  y_datos, 'o', label='datos medidos')
plt.plot(x_datos, y_exacto, '-', label='curva exacta')
plt.legend(loc='best')
plt.xlabel('x')
plt.ylabel('y')
plt.tight_layout()
# Además, vamos a graficar la relación entre las dos variables
plt.figure(figsize=(5, 2))
plt.scatter(x, y, color='red', alpha=0.5)
plt.title('Relación entre X e Y (no lineal)')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
# Vamos a asignar rangos y calcular Spearman usando la fórmula basada en rangos y diferencia de rangos

# Definimos la función para calcular Spearman basada en rangos y diferencia de rangos
def spearman_rango(x, y):
    # Primero, asignamos rangos a x e y, usando argsort de numpy para obtener los índices ordenados
    rangos_x = np.argsort(np.argsort(x))
    rangos_y = np.argsort(np.argsort(y))
    
    # Calculamos la diferencia de rangos
    d = rangos_x - rangos_y
    
    # Calculamos el coeficiente de Spearman
    n = len(x)
    return 1 - (6 * np.sum(d**2)) / (n * (n**2 - 1))
```

```python
# Calculamos Spearman para nuestras variables x e y
spearman = spearman_rango(x, y)

# Imprimimos Spearman
print(f"El coeficiente de Spearman entre X e Y es: {spearman:.2f}")
```

# EJERCICIO 9 – Correlación de Kendall desde cero
# -------------------------------------------------------------------
Instrucciones:

Simula dos variables no correlacionadas linealmente.
Asigna rangos y calcula Kendall

```python
# Vamos a simular dos variables no correlacionadas linealmente

# Definimos el tamaño de las variables
n = 100

# Generamos dos conjuntos de datos independientes con distribución normal usando numpy
x = np.random.normal(0, 1, n)
y = np.random.normal(0, 1, n)

# Como nos piden que las variables no deben estar correlacionadas,
# no realizamos ninguna transformación que las relacione linealmente.

# Vamos a graficar la relación entre las dos variables
plt.figure(figsize=(5, 2))
plt.scatter(x, y, color='green', alpha=0.5)
plt.title('Relación entre X e Y (no correlacionadas)')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
# Vamos a asignar rangos y calcular Kendall

# Definimos una función que calcula Kendall basada en rangos
def kendall_rango(x, y):
    # Primero, asignamos rangos a x e y, usando argsort de numpy para obtener los índices ordenados
    rangos_x = np.argsort(np.argsort(x))
    rangos_y = np.argsort(np.argsort(y))
    
    # Calculamos la diferencia de rangos
    d = rangos_x - rangos_y
    
    # Inicializamos las variables antes del bucle
    concordancias = 0
    discordancias = 0
    # Debemos calcular el número de concordancias y discordancias
    for i in range(len(d)):
        for j in range(i + 1, len(d)):
            if d[i] - d[j] < 0:
                concordancias += 1
            elif d[i] - d[j] > 0:
                discordancias += 1    
    # Calculamos el coeficiente de Kendall
    n = len(x)
    return (concordancias - discordancias) / (n * (n - 1) / 2)
```

```python
# Calculamos Kendall para nuestras variables x e y
kendall = kendall_rango(x, y)

# Imprimimos Kendall
print(f"El coeficiente de Kendall entre X e Y es: {kendall:.2f}")
```
